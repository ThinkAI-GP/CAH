iris est un jeu de données généralement utilisé pour l'apprentissage supervisé. Le but ici est de l'utiliser pour l'apprentissage non supervisé grace a la classification ascendante hiérarchique et de voir les regroupements de classes.

```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load in 

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
from scipy.cluster import hierarchy
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn import datasets
from sklearn.metrics import adjusted_rand_score
import seaborn as sns
```

```python
# import some data to play with
iris = datasets.load_iris()
```

```python
dir(iris)
```

```python
print("Les classes possibles (en apprentissage supervisé) sont {}".format(iris.target_names))
```

```python
iris.feature_names
```

```python
iris_df = pd.DataFrame(iris['data'], columns=iris['feature_names'])
iris_df["species"] =iris.target_names[iris.target]
sns.pairplot(iris_df,hue="species")
plt.show()
```

```python
#matérialisation des 3 classes (hauteur t = 7)
plt.figure(figsize=(10,10))
plt.title('CAH avec matérialisation des 3 classes')
Z = linkage(iris.data,method='ward',metric='euclidean')
dendrogram(Z,color_threshold=10)
plt.show()
```

```python
labels = hierarchy.fcluster(Z, t = 10, criterion = "distance")
print("Classes = {}".format(set(labels)))
print("Indice de Rand = {}".format(adjusted_rand_score(iris.target, labels)))
```
